## Handling missing values in listings

```python
# We know that each listing in the calendar file is associated with a unique price.

# We also know that 7313 listings in the listings file have a missing value in price.

# If the listings that have a missing price in the listings file were present in the calendar file then we would be able to retrieve the missing price.
# Which may not be precisely the updated price but may be a good estimate.
```

```python
# How many listings in the listings file with missing price are in the calendar file (with a non-missing price)?
(listings[listings['price_dkk'].isnull()]['id']).isin(calendar['listing_id']).shape[0]

#-> We are able to estimate all 7313 missing prices of listings file!
```

```python
# if of listings in the listings file with missing price
missing_price_ids = listings[listings['price_dkk'].isnull()]['id']

# looking for the corresponding prices in the calendar file
prices_from_calendar = calendar[calendar['listing_id'].isin(missing_price_ids)][['listing_id', 'price_dkk']]
prices_from_calendar = prices_from_calendar.drop_duplicates()

#  create a dictionary from prices_from_calendar
prices_dict = dict(zip(prices_from_calendar['listing_id'], prices_from_calendar['price_dkk']))
# fill missing values in listings['price_dkk'] with prices found in calendar['price_dkk]
listings['price_dkk'] = listings.apply(
    lambda row: prices_dict.get(row['id'], row['price_dkk']) if pd.isnull(row['price_dkk']) else row['price_dkk'], axis=1
)

# check the result
missing_after_fill = listings[listings['price_dkk'].isnull()]
missing_after_fill.shape[0]
#-> No more missing values in listings['price_dkk']
```
